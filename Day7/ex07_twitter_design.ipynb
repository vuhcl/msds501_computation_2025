### Design Twitter 
Design a simplified version of Twitter where users can post tweets, follow/unfollow another user, and is able to see the 10 most recent tweets in the user's news feed.

Implement the Twitter class:

- Twitter() Initializes your twitter object.
- postTweet(int userId, int tweetId) Composes a new tweet with ID tweetId by the user userId. Each call to this function will be made with a unique tweetId.
- getNewsFeed(int userId) Retrieves the 10 most recent tweet IDs in the user's news feed. Each item in the news feed must be posted by users who the user followed or by the user themself. Tweets must be ordered from most recent to least recent.
- follow(int followerId, int followeeId) The user with ID followerId started following the user with ID followeeId.
- unfollow(int followerId, int followeeId) The user with ID followerId started unfollowing the user with ID followeeId.

```python
class Twitter(object):
    def __init__(self):
        self.userPosting = dict() # userId : [tweetIds....] # add new at the end
        self.userFollowing = dict() # userId : {followeeIds .....}
        self.postCount = 0 # acts as a timestamp to check the posting sequence
    def postTweet(self, userId, tweetId):
        """
        Composes a new tweet with ID tweetId by the user userId. 
        Each call to this function will be made with a unique tweetId.
        """
        if userId in self.userPosting.keys():
            self.userPosting[userId].append((self.postCount, tweetId))
        else:
            self.userPosting[userId] = [(self.postCount, tweetId)]
        self.postCount += 1

    def getNewsFeed(self, userId):
        """
        Retrieves the 10 most recent tweet IDs in the user's news feed.
        Each item in the news feed must be posted by users who the user followed or by the user themself. 
        Tweets must be ordered from most recent to least recent.
        """
        poster_sets = set([userId])
       
        if userId in self.userFollowing.keys():
            for id in self.userFollowing[userId]:
                poster_sets.add(id)

        newsfeed = list()
        for poster in poster_sets:
            if poster in self.userPosting.keys():
                # Since each user's posting was appended on a list
                # retreiving the last 10 elements are all oredered by time
                newsfeed += self.userPosting.get(poster)[-10:]
        # Sort by time in descending order
        newsfeed.sort(key=lambda x : x[0], reverse=True)   
        return [ tweet[1] for tweet in newsfeed[:10] ]

    def follow(self, followerId, followeeId):
        """
        The user with ID followerId started following the user with ID followeeId.
        """
        if(followerId != followeeId):
            if followerId not in self.userFollowing.keys():
                self.userFollowing[followerId] = {followeeId}
            else:
                self.userFollowing[followerId].add(followeeId)

    def unfollow(self, followerId, followeeId):
        """
        The user with ID followerId started unfollowing the user with ID followeeId.
        """
        if followerId in self.userFollowing.keys():
            self.userFollowing[followerId].remove(followeeId)

# Your Twitter object will be instantiated and called as such:
# obj = Twitter()
# obj.postTweet(userId,tweetId)
# param_2 = obj.getNewsFeed(userId)
# obj.follow(followerId,followeeId)
# obj.unfollow(followerId,followeeId)
```
